```python
import pandas as pd
from sklearn import metrics
from sklearn.preprocessing import MultiLabelBinarizer
```

```python
df = pd.read_csv("Prediction.csv")
df['Predicted'] = df[df.columns.difference(['Category','Id'])].idxmax(axis=1)
```

```python
sum(df['Predicted']==df['C'])/len(df)
```

```python
prob_columns = list(df.columns.difference(['Category','Id','Predicted']))
```

```python
mlb = MultiLabelBinarizer()
label_matrix = mlb.fit_transform(df['Category'])
```
