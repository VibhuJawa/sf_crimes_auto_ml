```python
y = df_train['Category']
x = df_train[df_train.columns.difference(['Category'])]
```

```python
split_f = StratifiedShuffleSplit(1,test_size=0.05,train_size=0.95)
```

```python
for train_index, test_index in split_f.split(x, y):
    x_train,y_train = x.iloc[train_index],y.iloc[train_index]
    x_test,y_test = x.iloc[test_index],y.iloc[test_index]
```

```python
train_data = x_train.join(y_train)
test_data = x_test.join(y_test)
```

```python
train_data.to_csv('train_processed_split.csv',index=False)
```

```python
test_data.to_csv('test_processed_split.csv',index=False)
```
